```python
import os
import fnmatch
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def list_times_csv_files(directory):
    pattern = 'times*.csv'
    matching_files = []
    for root, dirs, files in os.walk(directory):
        for filename in fnmatch.filter(files, pattern):
            matching_files.append(os.path.join(root, filename))
    return matching_files

def extract_parameters_from_filename(filename):
    regex = r'times_(\w+)-k(\d+)-m(\d+)-ef(\d+)-n(\d+)-nq(\d+)-rep(\d+)\.csv'
    match = re.search(regex, filename)
    if match:
        experiment_type, k, m, ef, n, nq, rep = match.groups()
        return {
            'experiment_type': experiment_type,
            'k': int(k),
            'm': int(m),
            'ef': int(ef),
            'n': int(n),
            'nq': int(nq),
            'rep': int(rep)
        }
    return None

def process_csv_files(directory):
    times_csv_files = list_times_csv_files(directory)
    all_data = []

    for file in times_csv_files:
        parameters = extract_parameters_from_filename(file)
        if parameters:
            df = pd.read_csv(file)
            for key, value in parameters.items():
                df[key] = value
            all_data.append(df)

    data = pd.concat(all_data, ignore_index=True)
    data['time_per_query'] = data['total_time_ms'] / data['nq']
    return data

def plot_time_per_query(data):
    plt.figure(figsize=(16, 10))
    sns.set(style="whitegrid")

    # Define the order of experiment types
    experiment_order = ['cpu', 'batch', 'allocation', 'stream', 'pinned', 'cuda']

    # Separate data for batch and non-batch experiment types
    batch_data = data[data['experiment_type'] == 'batch']
    non_batch_data = data[data['experiment_type'] != 'batch']

    # Plot non-batch data with standard deviation
    sns.barplot(data=non_batch_data, x='experiment_type', y='time_per_query', hue='nq', ci='sd', order=experiment_order, palette='muted')

    # Plot batch data without standard deviation
    sns.barplot(data=batch_data, x='experiment_type', y='time_per_query', hue='nq', ci=None, order=experiment_order, palette='muted')

    # Manually handle the legend to avoid repetition
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys(), title='n_query', fontsize=12)

    plt.title('Average Time per Query by Experiment Type and n_query', fontsize=16)
    plt.xlabel('Experiment Type', fontsize=14)
    plt.ylabel('Average Time per Query (ms)', fontsize=14)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Directory containing the results
results_directory = '/content/hpgda_contest_MM/results'

# Process CSV files and extract data
data = process_csv_files(results_directory)

# Plot the time per query for all nq values
plot_time_per_query(data)
```

```python
def plot_time_per_query(data):
    # Filter for queries > 100
    filtered_data = data[data['nq'] >= 100]

    # Calculate mean time per query for each method, including CPU
    mean_times = filtered_data.groupby('experiment_type')['time_per_query'].mean()

    # Separate CPU value and other methods
    cpu_value = mean_times['cpu']
    gpu_times = mean_times.reindex(['batch', 'allocation', 'stream', 'pinned', 'cuda'])

    # Calculate percentage improvements
    improvements = []
    for i in range(len(gpu_times)-1):
        prev_val = gpu_times.iloc[i]
        curr_val = gpu_times.iloc[i+1]
        pct_improvement = ((prev_val - curr_val) / prev_val) * 100
        improvements.append(pct_improvement)

    plt.figure(figsize=(14, 8))
    sns.set_style("whitegrid")

    # Plot CPU reference line
    plt.axhline(y=cpu_value, color='r', linestyle='--', alpha=0.7)
    plt.text(
        -0.2, cpu_value,
        f'CPU: {cpu_value:.2e} ms',
        color='r',
        verticalalignment='bottom',
        bbox=dict(facecolor='white', edgecolor='red', alpha=0.7)
    )

    # Create line plot for GPU methods
    plt.plot(range(len(gpu_times)), gpu_times.values, 'bo-', linewidth=2, markersize=10)

    # Add method names on x-axis
    plt.xticks(range(len(gpu_times)), gpu_times.index, rotation=45)

    # Add percentage improvement labels between points
    for i in range(len(improvements)):
        x1, x2 = i, i+1
        y1, y2 = gpu_times.iloc[i], gpu_times.iloc[i+1]

        # Calculate position for the label
        mid_x = (x1 + x2) / 2
        mid_y = (y1 + y2) / 2

        # Add arrow and percentage label
        plt.annotate(
            f'-{improvements[i]:.1f}%',
            xy=(mid_x, mid_y),
            xytext=(0, 20),  # 20 points vertical offset
            textcoords='offset points',
            ha='center',
            va='bottom',
            bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0')
        )

    # Calculate and show total improvement from CPU to best GPU (CUDA)
    total_cpu_improvement = ((cpu_value - gpu_times.iloc[-1]) / cpu_value) * 100
    plt.text(
        len(gpu_times)-1, gpu_times.iloc[-1]+0.5,
        f'\nTotal improvement\nfrom CPU: -{total_cpu_improvement:.1f}%',
        ha='center',
        va='top',
        color='green',
        bbox=dict(facecolor='white', edgecolor='green', alpha=0.7)
    )

    plt.title('Performance Improvement Progression (Queries = 100)', fontsize=14)
    plt.xlabel('Method', fontsize=12)
    plt.ylabel('Average Time per Query (ms)', fontsize=12)

    # Format y-axis to use scientific notation for large numbers
    plt.gca().yaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
    plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))

    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

# Plot the time per query showing improvements
plot_time_per_query(data)
```

```python
import os
import fnmatch
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def list_recall_files(directory):
    pattern = 'log*.csv'  # Adjusted to match CSV files
    matching_files = []
    for root, dirs, files in os.walk(directory):
        for filename in fnmatch.filter(files, pattern):
            matching_files.append(os.path.join(root, filename))
    return matching_files

def extract_parameters_from_recall_filename(filename):
    regex = r'log_(\w+)-k(\d+)-m(\d+)-ef(\d+)-n(\d+)-nq(\d+)-rep(\d+)\.csv'
    match = re.search(regex, filename)
    if match:
        experiment_type, k, m, ef, n, nq, rep = match.groups()
        return {
            'experiment_type': experiment_type,
            'k': int(k),
            'm': int(m),
            'ef': int(ef),
            'n': int(n),
            'nq': int(nq),
            'rep': int(rep)
        }
    return None

def process_recall_files(directory):
    recall_files = list_recall_files(directory)
    all_recall_data = []

    for file in recall_files:
        parameters = extract_parameters_from_recall_filename(file)
        if parameters:
            df = pd.read_csv(file)  # Assuming CSV format
            for key, value in parameters.items():
                df[key] = value
            all_recall_data.append(df)
        else:
            print(f"Warning: Could not extract parameters from filename {file}")

    if all_recall_data:
        recall_data = pd.concat(all_recall_data, ignore_index=True)
        return recall_data

    return pd.DataFrame()  # Return an empty DataFrame if no data is found

def plot_recall_vs_k(recall_data):
    if 'k' not in recall_data.columns:
        print("Error: 'k' column is missing from the data.")
        return

    plt.figure(figsize=(14, 8))
    sns.set_style("whitegrid")

    # Create a line plot to show recall as k varies
    sns.lineplot(
        data=recall_data,
        x='k',
        y='recall',
        hue='experiment_type',
        marker='o',
        markersize=8,
        linewidth=2,
        palette='muted'
    )

    plt.title('Recall vs. k', fontsize=14)
    plt.xlabel('k', fontsize=12)
    plt.ylabel('Average Recall', fontsize=12)
    plt.legend(title='Method', fontsize=10, title_fontsize=12)

    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

# Example usage
results_directory = '/content/hpgda_contest_MM/results'
recall_data = process_recall_files(results_directory)

# Check if recall_data is empty or missing 'k'
if recall_data.empty:
    print("No recall data found.")
else:
    print(recall_data.head())  # Print the first few rows to verify the data

# Plot recall vs k
plot_recall_vs_k(recall_data)
```
