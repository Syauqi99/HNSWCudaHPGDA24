```python
# prompt: You see i have results folder i want you to compare the results of resultcuda with result_cpu. Remarks the filename is mostly the same except the cpu is using result_cpu wile Cuda is resultcuda. so to compare the cpu and cuda the filename of result_cpu or resultcuda should have the same name example result_cpu-k100-m16-ef100-n1000-nq1-rep1.csv and /content/hpgda_contest_MM/results/resultcuda-k100-m16-ef100-n1000-nq1-rep1.csv.  To compare for each files i want you to check if for each of the query_id of the cuda files have all the data_id of the cpu_files. So you match the query_id of the cpu and cuda files first then check the data_id. do this for all files and compile it in one table by calculating the accuracy

import pandas as pd
import os
import re

def compare_results(cpu_dir, cuda_dir):
    """
    Compares CPU and CUDA results, checking if all data_ids in CPU files are present in corresponding CUDA files.

    Args:
        cpu_dir: Path to the directory containing CPU result files.
        cuda_dir: Path to the directory containing CUDA result files.

    Returns:
        A pandas DataFrame with accuracy results for each file.
    """

    results = []
    for filename in os.listdir(cpu_dir):
        if filename.startswith("result_cpu"):
            match = re.search(r"result_cpu-(.+)\.csv", filename)
            if match:
                base_name = match.group(1)
                cpu_file = os.path.join(cpu_dir, filename)
                cuda_file = os.path.join(cuda_dir, f"resultcuda-{base_name}.csv")

                if os.path.exists(cpu_file) and os.path.exists(cuda_file):
                    try:
                        cpu_df = pd.read_csv(cpu_file)
                        cuda_df = pd.read_csv(cuda_file)

                        # Group data by query_id
                        cpu_grouped = cpu_df.groupby("query_id")
                        cuda_grouped = cuda_df.groupby("query_id")

                        correct_predictions = 0
                        total_predictions = 0

                        for query_id, cpu_group in cpu_grouped:
                          if query_id in cuda_grouped.groups:
                            cuda_group = cuda_grouped.get_group(query_id)
                            common_data_ids = set(cpu_group["data_id"]).intersection(set(cuda_group["data_id"]))
                            correct_predictions += len(common_data_ids)
                            total_predictions += len(cpu_group["data_id"])

                        accuracy = correct_predictions / total_predictions if total_predictions > 0 else 0
                        results.append({"filename": filename, "accuracy": accuracy})
                    except pd.errors.EmptyDataError:
                        print(f"Warning: Empty file encountered: {filename}")
                        results.append({"filename": filename, "accuracy": 0})
                    except KeyError as e:
                        print(f"Warning: Key error in file {filename}: {e}")
                        results.append({"filename": filename, "accuracy": 0})
    return pd.DataFrame(results)

# Example usage (assuming results are in /content/hpgda_contest_MM/results)
cpu_results_dir = "/content/hpgda_contest_MM/results"
cuda_results_dir = "/content/hpgda_contest_MM/results"  #Same directory for this example
comparison_results = compare_results(cpu_results_dir, cuda_results_dir)
comparison_results
```

```python
import os
import fnmatch
import re
import pandas as pd
import matplotlib.pyplot as plt

def list_times_csv_files(directory):
    pattern = 'times*.csv'
    matching_files = []
    for root, dirs, files in os.walk(directory):
        for filename in fnmatch.filter(files, pattern):
            matching_files.append(os.path.join(root, filename))
    return matching_files

def extract_parameters_from_filename(filename):
    regex = r'times_(\w+)-k(\d+)-m(\d+)-ef(\d+)-n(\d+)-nq(\d+)-rep(\d+)\.csv'
    match = re.search(regex, filename)
    if match:
        experiment_type, k, m, ef, n, nq, rep = match.groups()
        return {
            'experiment_type': experiment_type,
            'k': int(k),
            'm': int(m),
            'ef': int(ef),
            'n': int(n),
            'nq': int(nq),
            'rep': int(rep)
        }
    return None

def process_csv_files(directory):
    times_csv_files = list_times_csv_files(directory)
    all_data = []

    for file in times_csv_files:
        parameters = extract_parameters_from_filename(file)
        if parameters:
            df = pd.read_csv(file)
            for key, value in parameters.items():
                df[key] = value
            all_data.append(df)

    return pd.concat(all_data, ignore_index=True)

def plot_average_time(data):
    # Calculate average time per query
    data['average_time'] = data['total_time_ms'] / data['nq']

    # Group by experiment type and n_query, then calculate mean average time
    avg_time_data = data.groupby(['experiment_type', 'nq'])['average_time'].mean().reset_index()

    # Plot
    plt.figure(figsize=(12, 8))
    for experiment_type, group_data in avg_time_data.groupby('experiment_type'):
        plt.plot(group_data['nq'], group_data['average_time'], marker='o', label=experiment_type)

    plt.title('Average Time per Query by n_query and Experiment Type')
    plt.xlabel('n_query')
    plt.ylabel('Average Time per Query (ms)')
    plt.xscale('log')  # Use a logarithmic scale for better visualization
    plt.legend(title='Experiment Type')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Directory containing the results
results_directory = '/content/hpgda_contest_MM/results'

# Process CSV files and extract data
data = process_csv_files(results_directory)

# Plot the average time
plot_average_time(data)

```

```python
import os
import fnmatch
import re
import pandas as pd
import matplotlib.pyplot as plt

def list_times_csv_files(directory):
    pattern = 'times*.csv'
    matching_files = []
    for root, dirs, files in os.walk(directory):
        for filename in fnmatch.filter(files, pattern):
            matching_files.append(os.path.join(root, filename))
    return matching_files

def extract_parameters_from_filename(filename):
    regex = r'times_(\w+)-k(\d+)-m(\d+)-ef(\d+)-n(\d+)-nq(\d+)-rep(\d+)\.csv'
    match = re.search(regex, filename)
    if match:
        experiment_type, k, m, ef, n, nq, rep = match.groups()
        return {
            'experiment_type': experiment_type,
            'k': int(k),
            'm': int(m),
            'ef': int(ef),
            'n': int(n),
            'nq': int(nq),
            'rep': int(rep)
        }
    return None

def process_csv_files(directory):
    times_csv_files = list_times_csv_files(directory)
    all_data = []

    for file in times_csv_files:
        parameters = extract_parameters_from_filename(file)
        if parameters:
            df = pd.read_csv(file)
            for key, value in parameters.items():
                df[key] = value
            all_data.append(df)

    return pd.concat(all_data, ignore_index=True)

def plot_improvement(data):
    # Filter out data that is not of experiment type 'cpu'
    non_cpu_data = data[data['experiment_type'] != 'cpu']

    # Calculate average time per experiment type for non-cpu data
    avg_time_data = non_cpu_data.groupby('experiment_type')['total_time_ms'].mean().reset_index()

    # Define custom order
    custom_order = ['batch', 'allocation', 'stream', 'pinned', 'cuda']
    avg_time_data['order'] = avg_time_data['experiment_type'].apply(lambda x: custom_order.index(x))
    avg_time_data = avg_time_data.sort_values('order')

    # Calculate percentage improvement
    avg_time_data['improvement'] = avg_time_data['total_time_ms'].pct_change() * 100

    # Plot
    plt.figure(figsize=(12, 8))
    plt.plot(avg_time_data['experiment_type'], avg_time_data['total_time_ms'], marker='o', label='Average Time')
    for i in range(1, len(avg_time_data)):
        plt.annotate(f"{avg_time_data['improvement'].iloc[i]:.2f}%",
                     (i, avg_time_data['total_time_ms'].iloc[i]),
                     textcoords="offset points", xytext=(-15,10), ha='center', fontsize=9, color='black')

    # Add a red line for the average of "cpu" experiment type
    cpu_data = data[data['experiment_type'] == 'cpu']
    cpu_avg = cpu_data['total_time_ms'].mean()
    plt.axhline(y=cpu_avg, color='red', linestyle='--', label='CPU Average')

    plt.title('Average Total Time (ms) and Improvement by Experiment Type')
    plt.xlabel('Experiment Type')
    plt.ylabel('Average Total Time (ms)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Directory containing the results
results_directory = '/content/hpgda_contest_MM/results'

# Process CSV files and extract data
data = process_csv_files(results_directory)

# Plot the improvement
plot_improvement(data)
```
